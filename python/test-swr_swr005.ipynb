```python
%matplotlib inline
from IPython.display import Image
import os
import numpy as np
import matplotlib.pyplot as plt

import flopy


#Set the paths
datapth = os.path.join('..', 'test-swr', 'swr005')
```

### Load SWR Process observations

Create an instance of the `SwrObs` class and load the observation data.

```python
sobj = flopy.utils.SwrObs(os.path.join(datapth, 'swr005.obs'))

ts = sobj.get_data()

sobj0 = flopy.utils.SwrObs(os.path.join(datapth, '.cmp', 'swr005.obs.cmp'))
ts0 = sobj0.get_data()
```

#### Plot the data from the binary SWR Process observation file

```python
fig = plt.figure(figsize=(9, 9))
ax1 = fig.add_subplot(1, 1, 1)
ax1.plot(ts['totim'], ts['OBS1'], color='blue')
ax1.plot(ts['totim'], ts0['OBS1'], color='blue', lw=0, marker='o', ms=3)
ax1.plot(ts['totim'], ts['OBS2'], color='green')
ax1.plot(ts['totim'], ts0['OBS2'], color='green', lw=0, marker='o', ms=3)
ax1.plot(ts['totim'], ts['OBS3'], color='magenta')
ax1.plot(ts['totim'], ts0['OBS3'], color='magenta', lw=0, marker='o', ms=3)
ax1.plot(ts['totim'], ts['OBS4'], color='black')
ax1.plot(ts['totim'], ts0['OBS4'], color='black', lw=0, marker='o', ms=3)
ax1.set_ylabel('Stage, in meters')
#ax1.legend()
```

```python
for name in ts.dtype.names:
    d = ts[name] - ts0[name]
    print(name, d.min(), d.max(), d.mean())
```
